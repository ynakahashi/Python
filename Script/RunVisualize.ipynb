```python
x1 = np.random.randn(80)
x2 = np.random.randn(80)
x3 = x1 * x2
y1 = .5 + 2 * x1 - x2 + 2.5 * x3 + np.random.randn(80)
y2 = .5 + 2 * x1 - x2 + np.random.randn(80)
y3 = y2 + np.random.randn(80)
```

```python
sns.violinplot([x1, x2, x3])
```

```python
sns.regplot(x2, y2)
```

```python
df = pd.DataFrame(dict(x1=x1, x2=x2, x3=x3, 
                       y1=y1, y2=y2, y3=y3))
```
